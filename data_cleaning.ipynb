## The first bit of the program is for importing the required packages

> Pandas for dataframe and numpy for calculations

```python
import pandas as pd
import numpy as np
```

## Reading the specific sheet from the excel document
> Converting to dataframe

```python
df = pd.read_excel(r"Master_Data.xlsx")
data = pd.DataFrame(df)
data #Displays the dataframe
```

## Selecting the columns in which we will be using the dropna function
> * Normally, the dropna function will take the whole row (including the gene id), since gene id cannot be a NaN value and we need to remove all NaN value from the rows (excluding the genes), we specify the locus with the given code.
> * The given index specifies that we will be skipping the first column (i.e. the Gene_id column) and taking the rest.

```python
locus = data.columns[2:]
locus
```

## Replacing zeroes with NaN and dropping the rows with all null values
> * The inplace argument is used to commit the changes into the dataframe, by default, its a boolean set to False, here we're setting it to true so that the changes are committed to the dataframe. 
> * We use dropna to drop the null values, subset denotes the range in which the code will run (in this case, eerywhere excluding the Gene_id) and the how argument is used to specify that we want the entire row to be empty before we get it removed

```python
data.replace(0, np.nan, inplace=True)
data.dropna(subset=locus, how = "all", inplace = True)
data
```

## Replacing the NaN with zeroes again

```python
clean_data = data.fillna(0)
clean_data
```

## Save to new excel sheet

```python
clean_data.to_excel("CleanData.xlsx")
```
